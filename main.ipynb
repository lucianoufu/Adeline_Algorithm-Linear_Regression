# Adaline

```python
from random import randrange
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
t = [2.26, 3.80, 4.43, 5.91, 6.18, 7.26, 8.15, 9.14, 10.87, 11.58, 12.55]
w_ant = randrange(10)/10 -0.5
b_ant = 0
w_novo = w_ant
b_novo = b_ant
teta = 0
alfa = 0.01
erro_quadratico = 0
ciclos = 0
lista_ciclos = []
x_lista = []
y_lista = []
erro_lista = []
```

```python
while ciclos < 500:
    lista_ciclos.append(ciclos)
    ciclos += 1
    erro_quadratico = 0
    print(f"Ciclos {ciclos}")
    for entrada in range(11):
        y_liquido = w_ant * f[entrada] + b_ant
        y = y_liquido
        w_novo = w_ant + alfa * (t[entrada] - y) * f[entrada]
        b_novo = b_ant + alfa + (t[entrada] - y)
        w_ant = w_novo
        b_ant = b_novo
        erro_quadratico += (t[entrada] - y)**2
    erro_lista.append(erro_quadratico)
```

```python
for i in range(8):
    x_lista.append(i)
    y_lista.append(w_novo * x_lista[i] + b_novo)
```

```python
print(f'y = {w_novo} * x + {b_novo}' )
```

```python
fig = plt.figure(figsize = (20,20))
axes = fig.add_axes([0.5,0.5,0.5,0.5])
axes.plot(f, t, 'rs', x_lista, y_lista, 'b--')
plt.show()
```

```python
fig = plt.figure(figsize = (20,20))
axes = fig.add_axes([0.5,0.5,0.5,0.5])
axes.plot(lista_ciclos, erro_lista, 'rs')
plt.xlabel('Ciclos', fontsize = 20)
plt.ylabel('Erro quadratico', fontsize = 20)
plt.show()
```

<h1> Regresso linear</h1>

```python
x_array = np.array(f)
x2_array = x_array ** 2
y_array = np.array(t)
```

```python
print(x_array)
print(y_array)
```

```python
#a = len(x_array) * x2_array.sum() - x_array.sum()**2
a = ((len(x_array) * (x_array * y_array).sum()) - (x_array.sum() * y_array.sum())) / (len(x_array) * x2_array.sum() - x_array.sum()**2)
a
```

```python
b = y_array.mean() - (a * x_array.mean())
b
```

```python
print(f"y = {a} * x + {b}")
```

```python
x2_lista = []
y2_lista = []
for i in range(8):
    x2_lista.append(i)
    y2_lista.append(a * x_lista[i] + b)
x2_lista
y2_lista
```

```python
fig = plt.figure(figsize = (20,20))
axes = fig.add_axes([0.5,0.5,0.5,0.5])
axes.plot(f, t, 'rs', x2_lista, y2_lista, 'b--')
plt.show()
```

```python
print(f'Adaline: y = {w_novo} * x + {b_novo}')
print(f"Regresso linear: y = {a} * x + {b}")
```

```python
fig = plt.figure(figsize = (20,20))
axes = fig.add_axes([0.5, 0.5, 0.5, 0.5])
axes.plot(f, t, 'r^', x_lista, y_lista, 'b--', x2_lista, y2_lista, 'g--')

plt.show()
```
